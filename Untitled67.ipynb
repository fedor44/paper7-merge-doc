```python
from datasets import load_dataset
import pandas as pd

dataset1 = load_dataset('financial_phrasebank', 'sentences_75agree')
dataset2 = load_dataset('FinanceInc/auditor_sentiment')
```

```python
# pandas DataFrame으로 변환
df1 = pd.DataFrame(dataset1['train'])
df2 = pd.DataFrame(dataset2['train'])

# 'sentence' 컬럼을 기준으로 외부 조인 수행
merged_df = pd.merge(df2, df1, on='sentence', how='left', indicator=True)

# '_merge' 컬럼이 'left_only'인 데이터만 필터링 (즉, dataset2에만 있는 데이터)
unique_to_df2 = merged_df[merged_df['_merge'] == 'left_only']

# 필요 없는 '_merge' 컬럼 제거
unique_to_df2 = unique_to_df2.drop(columns=['_merge', 'label_y'])

# 'label_x' 컬럼의 이름을 'label'로 변경
unique_to_df2 = unique_to_df2.rename(columns={'label_x': 'label'})

unique_to_df2
```

```python
# 데이터셋 하나로 합치기
df1 = pd.read_csv('/content/drive/MyDrive/paper/financial_news_with_sentiment_1000.csv')
df2 = pd.read_csv('/content/drive/MyDrive/paper/auditor_sentiment_1100.csv')
```

```python
# 두개 데이터셋 모두 아래 조건에 label 맞춤
# 1(neutral), 2(positive), 0(negative)

# 두 데이터셋을 하나로 합치기
combined_df = pd.concat([df1, df2], ignore_index=True)
```
